## Build the HR from a FERMIBURST Query

First we create a HR builder object


```python
hrb = HardnessDurationBuilder()
```

Put in a query on the online FERMIBURST catalog parameters. E.g. t90>0 should be all GRBs

```python
hrb.catalog.query('.01<t90<5')
```

One we have our selection, we must go through and build the models. This routine selects the best fitting model (input 'fluence'[default] or 'peak') according the GBM online catalog. **This step can take a few minutes.**

```python
hrb.build_sample()
```

Now select the intervals for the HRs. Both a photon and an energy HR are calculted.

```python
hrb.compute_hardness_ratios(10,50,50,300)
```

We can view and access the data like so,

```python
hrb.hr_table
```
